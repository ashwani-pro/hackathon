```python
plt.figure(figsize=(15,10))
plt.scatter(data['mileage'].values, data['car_type'].values)
plt.plot(color='red')
plt.xlabel('Mileage')
plt.ylabel('Car Type')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(data['car_price'].values, data['engine_power'].values)
plt.plot(color='red')
plt.xlabel('Price')
plt.ylabel('Engine Power')
plt.grid(color='black')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['car_type'].values, data['fuel'].values)
plt.plot(color='red')
plt.xlabel('Car Type')
plt.ylabel('Fuel')
plt.show()
```
